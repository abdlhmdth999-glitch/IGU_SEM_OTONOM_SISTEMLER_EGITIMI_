```python
print("Programdan çıkmak için '0' girebilirsiniz.")

toplam = 0
s = 0

while True:
    sayi = int(input("Bir sayı girin: "))

    if sayi == 0:
        break

    toplam += sayi
    s += 1

if s == 0:
    print("Hiç sayı girişi yapmadınız!")
else:
    ortalama = toplam / s
    print(s, " adet sayı girildi.")
    print("Toplam: ", toplam)
    print("Ortalama: ", ortalama)
```

```python
ogrenciler = (
    ("Ali", (85, 90, 78)),
    ("Ayşe", (88, 76, 92)),
    ("Mehmet", (70, 85, 80)),
    ("Zeynep", (95, 90, 93))
)

for ogrenci in ogrenciler:
    isim = ogrenci[0]
    notlar = ogrenci[1]
    ortalama = sum(notlar) / len(notlar)
    print(isim, " adlı öğrencinin ortalaması: ", ortalama)
```
